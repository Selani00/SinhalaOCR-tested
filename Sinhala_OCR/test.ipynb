```python
from PIL import Image
from surya.ocr import run_ocr
from surya.model.detection.segformer import load_model as load_det_model
from surya.model.recognition.model import load_model as load_rec_model

image_path = "sinhala_text.jpg"
image = Image.open(image_path)

# Load models
det_model = load_det_model()
rec_model = load_rec_model()

# Run OCR (Surya automatically handles multilingual scripts including Sinhala)
predictions = run_ocr([image], [det_model], [rec_model])
print(predictions)
```
