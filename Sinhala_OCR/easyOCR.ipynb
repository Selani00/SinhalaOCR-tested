```python
import easyocr
from PIL import Image
import matplotlib.pyplot as plt

# Initialize OCR reader
reader = easyocr.Reader(['en'])  # Sinhala language

# Image path
image_path = "english_text.png"

# Extract text
results = reader.readtext(image_path)

# Print extracted text
print("Extracted Text:")
for item in results:
    print(item[1])

# Display image
img = Image.open(image_path)

plt.imshow(img)
plt.axis("off")
plt.show()
```
